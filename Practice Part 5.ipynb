```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
#styling argument
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
#styling argument
sns.set_style('white') #background white(default)
sns.countplot(x='sex',data=tips)
sns.despine() #remove top and right border
```

```python
#styling argument
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(bottom=True,right=False)
```

```python
#styling argument
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
#matplotlib set the size of seaborn plot
plt.figure(figsize=(15,8)) #12 width 3 height
sns.countplot(x='sex',data=tips)
```

```python
#scale and context
sns.set_context('poster', font_scale=1)
sns.countplot(x='sex',data=tips, palette='cool')
```
