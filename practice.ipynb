```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
# 2 ways to create plot 
# 1 functional and 2 oop bases
#FUNTIONAL
plt.plot(x,y) #if you are in jupyter notebook this is the way to show plot
plt.xlabel('X Side ka label')
plt.ylabel('Y side ka label')
plt.title('Title Bro')
```

```python
#if you arenot in jupyter notebook then this is the way to show plot
#plt.show() -> will not show as output if you arenot using jupyter note book you have to keep typing plt.show() 
#in order to print plot
```

```python
#creating Multiple Plots on same canvas
plt.subplot(1,2,1) # 1 -> row | 2-> column | 1-> plotnumber
plt.plot(x,y,'r')# r-> red

plt.subplot(1,2,2)# 1 -> row | 2-> column | 2-> plotnumber
plt.plot(y,x,'b')# b-> blue
```

```python
#Object Oriented Method
fig = plt.figure() #imaginary blank canvas fig-> onject
#add axis for canvas
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height
#plot on the axes
axes.plot(x,y)
#this is the same plot we drew using function plt.plot

#add x and y labels using set
axes.set_xlabel('OOP X label')
axes.set_ylabel('OOP Y label')
axes.set_title('Title')
```

```python
#plot 2 different plots using OOP
fig=plt.figure()
#add axis to show data later on on these axis
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #left,bottom,width,height

#plot on both axis(add data)
axes1.plot(x,y,'r')
axes2.plot(y,x,'b')

axes2.set_title('Smaller Plot')
axes1.set_title('Larger Plot')
```

```python
#again
fig=plt.figure()
#add axis to show data later on on these axis
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y,'r')
```

```python
#create a subplot using the same object oriented method
#plt.subplots instead of plt.figure doing add_axes autonatically instead of manual arguments 0.1 0.1 0.8 0.8
fig,axes=plt.subplots(nrows=1, ncols=2) # no of rows and columns decide the no of plots 1 row 2 col means 1 row p 2 plots 2 col me
#axis[0] and axis[1] are the two plots 
#lets plot x and y values and set titles and labels

axes[0].plot(x,y)
axes[0].set_xlabel('Left X Plot')
axes[0].set_ylabel('Left Y Plot')
axes[0].set_title('Left Title')


axes[1].plot(y,x)
axes[1].set_xlabel('Right X Plot')
axes[1].set_ylabel('Right Y Plot')
axes[1].set_title('Right Title')

plt.tight_layout() #overlapping issues resolve krne k lia
```

```python
#Figure Size, Aspect Ratio and DPI + label(legend-> 2 steps)
```

```python
fig= plt.figure(figsize=(12,2)) #width,height
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X Value')
ax.plot(x,x**2,label='X Square')
ax.plot(x,x**3,label='X Cube')



#you can add labels(legend) to each plot/line
ax.legend()
```

```python
fig,axes= plt.subplots(nrows=2,ncols=1,figsize=(8,2)) #width,height
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()


```
