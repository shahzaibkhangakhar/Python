# Regression plots

```python
#LM(linear model) plot function
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips ,hue='sex', markers=['o','v'], scatter_kws={'s':50}) #s-> size
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex' ) #seperated  by column
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time' ) #seperated  by column
```

```python
 sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time', aspect=1.7 , height=8 )
```
